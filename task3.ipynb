```python
# Import Librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import sklearn as sk
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/git-dibwar/DataScienceAndBusinessAnalytics/main/data/SampleSuperstore.csv')
```

```python
# Checking null values in the dataset
df.isnull().sum()
```

```python
# Drop the duplicate values
df.drop_duplicates(inplace=True)
```

```python
df.nunique()
```

```python
print('Total sales:', df["Sales"].sum())
```

```python
print('Total profit:', df["Profit"].sum())
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(),annot=True,cmap="Reds")
plt.show()
```

```python
sns.lineplot(x="Discount",y="Profit",data=df, color='red')
plt.show()
```

here we notice that the most profitable discount is 0 , 0.1 and 0.2

```python
sns.catplot(x='Category',y='Profit',data=df,kind='bar', color='pink')
```

the technology category has the highest profit between all the categories

```python
plt.figure(figsize=(15,15))
sns.countplot(x='Sub-Category',data=df)
plt.xticks(rotation=90,fontsize=17)
plt.title('Sales distribution per sub-category')
plt.show()
```

the top 3 best seller sub categories are : 
1- Binders
2- Paper 
3- Furnishings

```python
sns.catplot(x='Ship Mode',y='Profit',data=df,kind='bar')
```

 The profits of the different ship modes do not differ much.

```python
plt.figure(figsize=(14,14))
sns.countplot(x='State',data=df)
plt.xticks(rotation=90,fontsize=16)
plt.title('Sales distributed according to States')
plt.show()
```

Most customers are located in California, New York and Texas, so these states have got to be more profitable than the others.
